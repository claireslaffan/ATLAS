```python
def fit_hist(pdf, pdf_name, func):
    print("number of entries: " + str(len(pdf)))
    bins=np.linspace(0,0.35,50)
    fig, ax = plt.subplots(figsize=(10,8))
    bin_heights, bin_borders, _ = plt.hist(pdf, bins=bins, density=False, label='histogram')

    bin_centers = bin_borders[:-1] + np.diff(bin_borders) / 2
    print("bin width: " + str(np.diff(bin_borders)[0]))
    
    popt, _ = curve_fit(func, bin_centers, bin_heights, maxfev = 1000000)

    x_interval_for_fit = np.linspace(bin_borders[0], bin_borders[-1], 1000)
    ax.plot(x_interval_for_fit, func(x_interval_for_fit, *popt), linewidth=2, label='fit')
    ax.set_xlabel(pdf_name,fontsize=30)

    ax.tick_params(labelsize=16)
    ax.legend(fontsize=16)
    
    return popt
```

```python
def find_slices(hist, pT, desired_bins):
    
    threshold = np.ceil(len(hist)/desired_bins) 
    previous = 0
    ranges = [[0,0]]
    #step through pT in order, and see how many hist events you get until reach the threshold
    for vals in np.sort(pT):
        eggs = len(hist[(pT>previous)&(pT<vals)])
        #print(eggs)
        if (eggs == threshold) or (vals == np.sort(pT)[len(pT)-1]):
            ranges.append([previous, vals]) #this is your pT cut 
            previous = vals
            print('num events: ' + str(eggs))
    return ranges    
```

```python
def mean (weight, amt):
    num = 0
    mean = 0
    for i in range(len(weight)):
        num += weight[i]*amt[i]
    
    mean = num/len(weight)
    return mean
    
```
